```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv(r"C:\Users\tatak\haroon\DATASETS\Climate.csv")
df.head()
```

```python
df.drop(["Day"],axis=1,inplace=True)
```

```python
dfdict=pd.DataFrame(df.dtypes)
dfdict['MissingVal'] =df.isnull().sum()
dfdict['NUnique']=df.nunique()
dfdict['Count']=df.count()
dfdict=dfdict.rename(columns={0:'DataType'})
dfdict
```

```python
df1=df.select_dtypes(include="number")
```

```python
df1.boxplot()
plt.xticks(rotation=70)
plt.show()
```

```python
for i in df1:
    Q1=df[i].quantile(.25)
    Q3=df[i].quantile(.75)
    IQR=Q3-Q1
    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR
    df[i]=df[i].clip(lower_limit,upper_limit)
```

```python
df.boxplot()
plt.xticks(rotation=70)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler,MaxAbsScaler
```

```python
df.columns
```

```python
scaler=StandardScaler()
cols=['Average humidity (%)',
       'Average dewpoint (°F)', 'Average barometer (in)',
       'Average windspeed (mph)', 'Average gustspeed (mph)',
       'Average direction (°deg)', 'Rainfall for month (in)',
       'Rainfall for year (in)', 'Maximum rain per minute ',
       'Maximum temperature (°F)', 'Minimum temperature (°F)',
       'Maximum humidity (%)', 'Minimum humidity (%)', 'Maximum pressure ',
       'Minimum pressure ', 'Maximum windspeed (mph)',
       'Maximum gust speed (mph)', 'Maximum heat index (°F)']
sc=scaler.fit_transform(df[cols])
df[['Average humidity (%)',
       'Average dewpoint (°F)', 'Average barometer (in)',
       'Average windspeed (mph)', 'Average gustspeed (mph)',
       'Average direction (°deg)', 'Rainfall for month (in)',
       'Rainfall for year (in)', 'Maximum rain per minute ',
       'Maximum temperature (°F)', 'Minimum temperature (°F)',
       'Maximum humidity (%)', 'Minimum humidity (%)', 'Maximum pressure ',
       'Minimum pressure ', 'Maximum windspeed (mph)',
       'Maximum gust speed (mph)', 'Maximum heat index (°F)']]=pd.DataFrame(sc,columns=cols)
df
```

```python
x=df.drop(["Average temperature (°F)"],axis=1)
x
```

```python
y=df["Average temperature (°F)"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor

from sklearn.metrics import r2_score
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
y_predlm=lm.predict(x_test)
r2lm=r2_score(y_test,y_predlm)
r2lm
```

```python
rd=Ridge(alpha=1.0)
rd.fit(x_train,y_train)
```

```python
y_predrd=rd.predict(x_test)
r2rd=r2_score(y_test,y_predrd)
r2rd
```

```python
ls=Lasso(alpha=0.1)
ls.fit(x_train,y_train)
```

```python
y_predls=ls.predict(x_test)
r2ls=r2_score(y_test,y_predls)
r2ls
```

```python
rf=RandomForestRegressor(n_estimators=100)
rf.fit(x_train,y_train)
```

```python
y_predrf=rf.predict(x_test)
r2rf=r2_score(y_test,y_predrf)
r2rf
```

```python
dt=DecisionTreeRegressor(criterion="absolute_error",random_state=42)
dt.fit(x_train,y_train)
```

```python
y_preddt=dt.predict(x_test)
r2dt=r2_score(y_test,y_preddt)
r2dt
```

```python
svrl=SVR(kernel="linear")
svrl.fit(x_train,y_train)
```

```python
y_predsvrl=svrl.predict(x_test)
r2svrl=r2_score(y_test,y_predsvrl)
r2svrl
```

```python
svrr=SVR(kernel="rbf")
svrr.fit(x_train,y_train)
```

```python
y_predsvrr=svrr.predict(x_test)
r2svrr=r2_score(y_test,y_predsvrr)
r2svrr
```

```python
metric_k=[]
neighbors=np.arange(3,15)
```

```python
for i in neighbors:
    knr=KNeighborsRegressor(n_neighbors=i,metric="euclidean")
    knr.fit(x_train,y_train)
    y_predknr=knr.predict(x_test)
    r2knr=r2_score(y_test,y_predknr)
    metric_k.append(r2knr)
```

```python
plt.plot(neighbors, metric_k,'o-')
plt.xlabel("neighbors")
plt.ylabel("accuracy")
plt.grid()
plt.show()
```

```python
knr=KNeighborsRegressor(n_neighbors=9,metric="euclidean")
knr.fit(x_train,y_train)
y_predknr=knr.predict(x_test)
r2knr=r2_score(y_test,y_predknr)
metric_k.append(r2knr)
```
